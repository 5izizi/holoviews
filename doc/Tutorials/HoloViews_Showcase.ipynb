# HoloViews Showcase

**Holoviews encapsulates your data into storable, composable objects, leveraging matplotlib for visualization and IPython Notebook for maximum productivity.**

Main features:

* Preserve your raw data while engaging in rapid, interactive exploratory analysis.
* Build complex, composite visualizations that are easily interactively customized at any time.
* Short, declarative style enables all necessary code to be presented, enabling your workflow is fully reproducibile.
* Select, slice or transform only the relevant data that is currently of interest.
* Store your raw data as holoviews objects via pickling.
* Strong support for the IPython Notebook including tab-completion throughout and powerful magics.
* Seamless interaction with Pandas Dataframes.

Using IPython notebook is only the first step towards an interactive, exploratory workflow using Python's rich ecosystem of data analysis tools. Holoviews will greatly improve your productivity, requiring fewer lines of code, allowing you to spend more time to explore your data instead of writing code to visualize and process it.

```python
from holoviews import *
from holoviews.operation import contours, threshold, gradient
import numpy as np
```

For this showcase, we use an unqualified import but in general we recommend using qualified imports and namespaces.

# A simple function

This is the function we will explore:

```python
def sine(x, multiplier=1, phase=0, freq=100):
    return multiplier * np.sin((freq * x + phase))
```

We will explore the effect of varying amplitude, phase and frequency:

```python
multipliers =  np.linspace(0.5,2,2) # Explored multiplier values
phases = np.linspace(0,2*np.pi,6)  # Explored phases
freqs = np.linspace(50,150,5)       # Explored frequencies
```

We will need some spatial sampling too:

```python
dist = np.linspace(-0.5,0.5,202)   # Linear spatial sampling
x,y = np.meshgrid(dist, dist)
grid = (x**2+y**2)                 # Spatial sampling (2D)
```

# Compact data visualization

With holoviews, we can immediately test our simple function by creating ``Matrix`` and ``Curve`` objects to visualize the 2D-arrays and 1D cross-sections generated by the ``sine`` function respectively:

```python
freq1 = Matrix(sine(grid, freq=50)) + Curve(zip(dist, sine(dist**2, freq=50)))
freq2 = Matrix(sine(grid, freq=200)) + Curve(zip(dist, sine(dist**2, freq=200)))
(freq1 + freq2).cols(2)
```

As you can see, ``Matrix`` takes a 2D numpy array as input and ``Curve`` accepts a list of ``(x,y)`` points. With the ``+`` operator we can lay these elements out together and with ``.cols(2)`` we can generate a 2 column format.

Only two element types are shown above but Holoviews supports many other types of element that behave in the same way: **scatter points, histograms, tables, vectorfields, RGB images, 3D plots, annotations** and ***many more*** as shown in [this tutorial]().

# Interactive exploration

To interactively explore this simple function, we need to declare the dimensions of the parameter space to be explored (``dimensions``) as well as the sample keys in this parameter space (``keys``):

```python
dimensions = ['Multiplier','Phase', 'Frequency']
keys = [(m,p,f) for p in phases for f in freqs for m in multipliers]
```

Now we create our high-dimensional ``HoloMap`` which can be explored: The tuple keys are the points in the parameter space and the values are the corresponding ``Matrix`` objects:

```python
items = [(k, Matrix(sine(grid, *k), value_dimensions=['Amplitude'])) for k in keys]
circular_wave = HoloMap(items, key_dimensions=dimensions)
circular_wave
```

You can still compose as many visualization elements as you wish together. Here is a demonstration of how to generate the horizontal cross-section of the circular wave using ``Curve`` elements. This is then positioned next to our circular wave:

```python
items = [(k, Curve(zip(dist, sine(dist**2, *k)))) for k in keys]
sections = HoloMap(items, key_dimensions=dimensions)
circular_wave + sections
```

Export your ``HoloMap`` objects to interactive notebook, video formats or gif.

# Succinct Analysis

Holoviews offers a suite of general operations to analyse your data, generating new HoloView structures in the process. Here we pick a point on the circular wave at ``(0,0.25)`` and plot the amplitude value as a function of phase.

The circular wave is shown annotated with a point at the chosen position. Note how we can still interactively explore the remaining dimensions, namely ``Multiplier`` and ``Frequency``.

```python
sample_pos = (0,0.25)
annotated = circular_wave * Points([sample_pos])
sample = circular_wave.sample(samples=[sample_pos]).to.curve('Phase', 'Amplitude')
annotated + sample
```

Note that the spatial frequency of our curve plot is *not* affected by the frequency of our wave. That is because the phase always spans exactly one cycle at any of the chosen frequencies.

Here is the our circular wave thresholded next to the same wave annotated with contours at the 0.5 level:

```python
%%view holomap='gif'
%%opts Contours (color='r')
m=HoloMap([(p, Matrix(sine(grid, phase=p))) for p in phases], key_dimensions=['Phase'])
contours(m, levels=[0.5]) + gradient(m) + threshold(m, level=0.5).hist()
```

Supplied operations: **gradient, fft_power, convolve, histogram, vectorfield, analyze_roi** and many more.

## Understand your data

Holoviews is designed to make it easy to understand your data. For instance, these two circular waves have two very different amplitudes. Holoviews ensures these two elements are normalized together when displayed together:

```python
%%opts Matrix (cmap='hot')
comparison = Matrix(sine(grid, multiplier=1)) + Matrix(sine(grid, multiplier=0.02, phase=np.pi))
comparison
```

Unfortunately, although we now have a clear representation of the relative amplitudes, the spatial structure of the low-amplitude wave is hard to see. Fortunately, it is easy to instruct holoviews to stop normalizing the group:

```python
%%opts Matrix {-groupwise} (cmap='hot')
comparison
```
